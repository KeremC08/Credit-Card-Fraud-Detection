```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data['Amount'] = scaler.fit_transform(pd.DataFrame(data['Amount']))
data.head()
```

```python
data.drop(['Time'],axis=1)
```

```python
data.duplicated().any()
```

```python
data = data.drop_duplicates()
data.shape
```

```python
data['Class'].value_counts()
```

```python
x = data.drop('Class', axis=1)
y = data['Class']
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=42)
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score

classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
}

for name,clf in classifier.items():
    print(f"\n==========={name}==========")
    clf.fit(train_x, train_y)
    y_pred = clf.predict(test_x)
    acc = accuracy_score(test_y, y_pred)
    precision = precision_score(test_y, y_pred)
    recall = recall_score(test_y, y_pred)
    f1 = f1_score(test_y, y_pred)
    print(f"Accuracy: {acc}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")
    
```

```python
#Undersampling

normal = data[data['Class'] == 0]
fraud = data[data['Class'] == 1]

normal.shape


```

```python
fraud.shape
```

```python
normal_sample = normal.sample(n=473)
normal_sample.shape 
```

```python
new_data = pd.concat([normal_sample,fraud],ignore_index=True)
new_data = new_data.drop(['Time'], axis=1)
new_data.head()
x  = new_data.drop('Class', axis=1)
```

```python
x  = new_data.drop('Class', axis=1)
y=new_data['Class']
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=42)
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score

classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}

for name,clf in classifier.items():
    print(f"\n==========={name}==========")
    clf.fit(train_x, train_y)
    y_pred = clf.predict(test_x)
    acc = accuracy_score(test_y, y_pred)
    precision = precision_score(test_y, y_pred)
    recall = recall_score(test_y, y_pred)
    f1 = f1_score(test_y, y_pred)
    print(f"Accuracy: {acc}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")
    
```

```python
#Oversampling

data = data.drop(['Time'], axis=1)
x = data.drop('Class', axis=1)
y = data['Class']

x.shape



```

```python
from imblearn.over_sampling import SMOTE
x_res,y_res = SMOTE().fit_resample(x, y)
y_res.value_counts()
```

```python


train_x, test_x, train_y, test_y = train_test_split(x_res, y_res, test_size=0.2, random_state=42)



classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
}


for name, clf in classifier.items():
    print(f"\n==========={name}==========")
    train_x.shape
    test_x.shape

    clf.fit(train_x, train_y)
    y_pred = clf.predict(test_x)
    acc = accuracy_score(test_y, y_pred)
    precision = precision_score(test_y, y_pred)
    recall = recall_score(test_y, y_pred)
    f1 = f1_score(test_y, y_pred)
    print(f"Accuracy: {acc}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")
    
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_res, y_res)
import joblib
joblib.dump(dtc,"models/creditcard_model2.pkl")

```
